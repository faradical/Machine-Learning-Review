```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

## Generating Synthetic Data
The "target" here is used just for demonstration purposes. In the real world, we would not have this.

```python
X, target = make_moons(n_samples=500, noise=0.1)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df = pd.DataFrame()
df['x'] = X_scaled[:,0]
df['y'] = X_scaled[:,1]

df.head()
```

## Plotting our data

```python
plt.scatter(df['x'], df['y'])
plt.show()
```

Oh no! The clusters in this data set appear to not be linearly separable. Previously in the course, we used K Means to separate clusters, but because K Means tries to draw a line between them, we will need to try something else.

There are many clustering alogorithms that can used to approach this problem. A great overview of these can be found [here](https://scikit-learn.org/stable/modules/clustering.html#clustering) on the sci-kit learn website. The following image provides excellent visual insight into how many of these clustering models approach problems.

![Clustering](https://scikit-learn.org/stable/_images/sphx_glr_plot_cluster_comparison_001.png)

For this notebook, we will continue by using the DBSCAN algorithm

## Clustering with DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X_scaled)
labels = dbscan.labels_
print(f"Possible labels: {set(labels)}")
labels
```

## Plotting the predicted clusters

Color coding our predicted clusters

```python
plt.scatter(df['x'], df['y'], c=labels)
plt.show()
```

Comparing our predictions to the target column we got from doing make moons. Again, we would not have this in the real world.

```python
plt.scatter(df['x'], df['y'], c=target)
plt.show()
```
