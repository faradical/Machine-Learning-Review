```python
from MyModule.Model import PipedModel, reshape_sample, reshape_feature
```

Making predictions on a single sample of data:

```python
PipedModel.predict(reshape_sample(200))
```

Plotting the model curve:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating predictons using loaded model
X = np.arange(222)
predictions = PipedModel.predict(reshape_feature(X))

# Plotting the dummy data and predictions together.
plt.figure(facecolor='gray', figsize=(15,10)).set_alpha(0.0)
ax = plt.axes()
ax.set_facecolor("gray")
ax.set_alpha(0.0)

plt.plot(X, predictions, c="black")
plt.show()
```
